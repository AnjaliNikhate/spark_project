##**- observation**


*   Country has only one value , United States; we will be dropping the country column.

```python
data = data.drop(columns=['Country','Postal Code'] , axis=1)
data.head()
```



*   We drop country column and postal code column because it does not directly affect profit and sales.



#**Data Visualization**

```python
#histogram distribution of numeric features
data.hist(bins=20,figsize=(18, 16),)
```

##**Heatmap**

```python
#correlation between features
corr = data.corr()
corr
```

```python
# visualizing correlation between features 
plt.figure(figsize=(15,9))
sns.heatmap(corr,annot=True,cmap="inferno")
plt.show()
```

##**observation**


*   From the heatmap , we can see here a positive cor-realtion between sales and profit that is 0.48 , thus profit increases, simultanously sales increases.
*   and we can also see here a negative cor-relation between profit and discount of -0.22 and sales and discount of -0.028 , thus discount increases, simultanously profit goes down.

```python
#calculating statewise profit
profit_state_g = data.groupby('State')['Profit'].sum()
profit_state_g
```

#**Statewise profit of US**

```python
#statewise profit of US
plt.subplots(figsize=(15,17),)
profit_state_g.plot(kind='barh', fontsize=15,color='darkslategrey')
plt.xlabel('profit', fontsize=15,)
plt.ylabel('state', fontsize=13)
plt.title('Statewise profit', fontsize=14)
plt.show()
```

##**- Observation**


*   Highest profit is generated by California and Newyork.
*   Lowest profit is generated by Texas and ohio

by the subplot we can clearly see that :


*   Top 5 states giving profit are **California** , **Newyork** , **Washington** and **michigan**.
*   Top 5 sates giving loss are **Texas** , **Ohio** , **Illinois** , **pennsylvania.**

```python
#calculating statewise sales
sales_state_g = data.groupby('State')['Sales'].sum()
sales_state_g
```

```python
# State wise Sales of US
plt.figure(figsize=(18,12))
sns.barplot(x=sales_state_g.index, y=sales_state_g.values ,palette='rocket')
plt.ylabel("Sales")
plt.xticks(rotation=80)
plt.show()
```

##**Observation**


*   Thus giving max profit California also has most sales. -we can see that the Lowest sales are from North Dakota.
        By the subplot we can signify that
*   Top 5 states giving most of sales are from **California** , **Newyork** , **Washington** and **Texas**.
*   Top 5 sates giving loss are **North Dakota** , **Maine** , **South Dakota**, **West Virginica**.

```python
#Comparing Sales and Profit Statewise
sales_profit = data.groupby('State')['Sales','Profit'].sum()
sales_profit.plot.bar(figsize=(14,9),)
plt.show()
```

##**Observation**


*   state California , Washington , Newyork ,michigan have highest sales and profit
*   state Texas , pennsylvania , Ohio , Illinois have higher sales, but profit in negative i.e. in loss

```python
data.Discount.head(9)
```

```python
#State wise avearge Discount of US
avg_dis = np.round(data.groupby('State')['Discount'].mean()*100,decimals=0)
avg_dis
```

```python
#State wise avg_discount of US
plt.figure(figsize=(15,9))
sns.barplot(x=avg_dis.index,y=avg_dis.values,palette='Set1')
plt.xticks(rotation=80)
plt.show()
```

##**Observation**


*   Illinois and Texas has highest average discount among other state of USA
*   we can also see that here Top 5 States having most discount are Illinois , Texas , Pennsylvania , Ohio and Colorado

```python
# Lineplot between Profit and Discount
plt.figure(figsize=(8,6))
sns.lineplot(x='Discount',y='Profit',data=data,color='green')
plt.show()
```

##**Observation**


*   By the help of lineplot we can see trend of Profit with respect to Discount.
*   Simultaneously Discount increases and Profit get decrease.

```python
# Lineplot between Sales and Discount
plt.figure(figsize=(8,6))
sns.lineplot(x='Discount',y='Sales',data=data,color='purple')
plt.show()
```

##**Observation**


*   By the help of lineplot we can see that trend of Sales with respect to Discount.

```python
#Calculating category wise sales , Discount , Profit
cat_data = data.groupby(['Category'])[['Sales','Discount','Profit']].mean()
cat_data
```

```python
#plotting donut chart to see category wise sales , Discount , Profit
ex= [0.05,0.05,0.05]
cat_data.plot.pie(subplots=True,figsize=(19,20),autopct='%0.1f%%',label=cat_data.index,
                  pctdistance=0.65,explode=ex)

plt.show()
```

##**Observation**
*   Technology category has the maximum Sales and profit and less discount compare to furniture and office.
*   Office supplies and Furniture category has minimum sales and profit as compare to Technology category.

```python
#Calculating Ship Mode wise sales , Discount , Profit
ship_data = data.groupby(['Ship Mode'])[['Sales','Discount','Profit']].mean()
ship_data
```

```python
#plotting donut chart to see category wise sales , Discount , Profit
ex= [0.05,0.05,0.05,0.05]
ship_data.plot.pie(subplots=True,figsize=(19,20),autopct='%0.1f%%',label=cat_data.index,
                  pctdistance=0.65,explode=ex)

plt.show()
```

##**Observation**
*   In Ship Mode, Same Day has the highest sales.
*   First Class has the highest Profit and Discount.
*   In Ship Mode, Standard Class has lowest sales as well as lowest profit.

```python
#Comparing Sales Sub-Category Wise
sub_cat = data.groupby('Sub-Category')['Sales'].mean()
# sub_cat
plt.figure(figsize=(15,9))
sns.barplot(x=sub_cat.index,y=sub_cat.values,palette='viridis')
plt.xticks(rotation=80)
plt.show()
```

##**Observation**
*   Sub-Category items Copier , Machines and Table have the highest Sales.
-copiers had highest sale ,Machines had second highest sales and Table had third highest sales.

```python
#Comparing  Profit Sub-Category Wise
sub_cat = data.groupby('Sub-Category')['Profit'].sum()
# sub_cat
plt.figure(figsize=(15,9))
sns.barplot(x=sub_cat.index,y=sub_cat.values,palette='icefire_r')
plt.xticks(rotation=80)
plt.show()
```

##**Observation**
*   we can see here Sub-Category, items Copiers has the highest Sales as well as highest profit margin.
*   and also we can see here Machine had second highest sales but less profit margin ,due to giving large amount of discount.
*   In sub-category, item Table had third highest sales but it's profit in negative i.e. in loss.
